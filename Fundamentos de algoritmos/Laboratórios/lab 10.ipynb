```python
#ex 1-Crie uma matriz M[12][12] com números inteiros aleatórios (de 0 até 10: randint(0,10)); Então, leia um caractere maiúsculo (S ou M), que indica uma operação que deve ser realizada na matriz. Em seguida, calcule e mostre a soma (S) ou a média (M) considerando somente aqueles elementos que estão abaixo da diagonal secundária da matriz, conforme ilustrado abaixo (área verde). Exiba a matriz criada e o resultado da operação.
#Para imprimir os elementos da matriz utilize "%3d" para dar espaço de 3 caracteres.
#Utilize somente número inteiros (inclusive no cálculo da média).

from random import randint

M = [[randint(0,10)for _ in range(12)]for _ in range(12)]

# Leitura da operação
operacao = input("")

# Exibindo a matriz
print("Matriz criada:")
for linha in M:
    for elemento in linha:
        print(f"{elemento:3d}", end=" ")
    print()



# Calculando a soma dos elementos abaixo da diagonal secundária
soma = 0
contador = 0
for i in range(12):
    for j in range(12):
        if i + j > 11:  # Condição para estar abaixo da diagonal secundária
            soma += M[i][j]
            contador += 1

# Realizando a operação
if operacao == 'S':
    print(f"Resultado da conta: {soma}")
elif operacao == 'M':
    media = soma // contador  # Média inteira
    print(f"Resultado da conta: {media}")
else:
    print("Operação inválida. Digite 'S' ou 'M'.")
```

```python
#ex 3- Escreva um algoritmo que leia um inteiro N (0 ≤ N ≤ 15), correspondente a ordem de uma matriz M de inteiros, e construa a matriz de acordo com o exemplo abaixo. Reserve 4 casas para imprimir cada dado da matriz.

n = int(input("Digite a ordem da matriz: "))
matriz = []
for i in range(n):
    linha = []
    atual = 2 ** i
    for _ in range(n):
        linha.append(atual)
        atual *= 2
    matriz.append(linha)

for linha in matriz:
    print("".join(f"{numero:4d}" for numero in linha)) #"".join(...) significa que os itens serão unidos sem nenhum separador (string vazia "" entre os itens).
```

```python
#ex 4-

# criar a matriz 2x2
matriz = []
for i in range(2):
    linha = []
    for j in range(2):
        valor = int(input(f"Digite o elemento de linha {i} e coluna {j}: "))
        linha.append(valor)
    matriz.append(linha)

# encontrar o maior elemento
maior = max(max(linha) for linha in matriz)

# calcular a matriz resultante
resultado = [[elemento * maior for elemento in linha] for linha in matriz]

# imprimir a matriz formatada
print("Matriz resultante:")
for linha in resultado:
    print(" ".join(f"{elemento:3d}" for elemento in linha))
```
