```python
#ex4
// Solicita ao usuário que insira um número inteiro
let numero = parseInt(prompt("Digite um número inteiro para calcular o fatorial:"));

// Função para calcular o fatorial
function calcularFatorial(n) {
    if (n === 0 || n === 1) {
        return 1;
    }
    let fatorial = 1;
    for (let i = 2; i <= n; i++) {
        fatorial *= i;
    }
    return fatorial;
}

// Verifica se o número é válido
if (isNaN(numero) || numero < 0) {
    console.log("Por favor, insira um número inteiro não negativo.");
} else {
    // Calcula e exibe o fatorial
    let resultado = calcularFatorial(numero);
    console.log(`O fatorial de ${numero} é ${resultado}.`);
}
```

```python
// Exemplo de dados coletados da pesquisa
let habitantes = [
    { idade: 25, sexo: 'F', salario: 3000 },
    { idade: 30, sexo: 'M', salario: 4000 },
    { idade: 22, sexo: 'F', salario: 2500 },
    { idade: 35, sexo: 'M', salario: 6000 },
    { idade: 28, sexo: 'F', salario: 5000 },
    { idade: 40, sexo: 'F', salario: 4500 }
];

// Inicialização das variáveis para cálculos
let totalSalarios = 0;
let maiorIdade = habitantes[0].idade;
let menorIdade = habitantes[0].idade;
let mulheresSalarioAte5000 = 0;

// Processamento dos dados
for (let i = 0; i < habitantes.length; i++) {
    let habitante = habitantes[i];

    // Cálculo da média de salário
    totalSalarios += habitante.salario;

    // Determinação da maior e menor idade
    if (habitante.idade > maiorIdade) {
        maiorIdade = habitante.idade;
    }
    if (habitante.idade < menorIdade) {
        menorIdade = habitante.idade;
    }

    // Contagem de mulheres com salário até R$5000,00
    if (habitante.sexo === 'F' && habitante.salario <= 5000) {
        mulheresSalarioAte5000++;
    }
}

// Cálculo da média de salário
let mediaSalario = totalSalarios / habitantes.length;

// Exibição dos resultados
console.log(`a. Média de salário do grupo: R$${mediaSalario.toFixed(2)}`);
console.log(`b. Maior idade do grupo: ${maiorIdade} anos`);
console.log(`b. Menor idade do grupo: ${menorIdade} anos`);
console.log(`c. Quantidade de mulheres com salário até R$5000,00: ${mulheresSalarioAte5000}`);
```

```python
#ex extra 1
const notaLaboratorio = parseFloat(prompt("Digite a nota do trabalho de laboratório:"));
const notaAvaliacaoSemestral = parseFloat(prompt("Digite a nota da avaliação semestral:"));
const notaExameFinal = parseFloat(prompt("Digite a nota do exame final:"));

const pesoLaboratorio = 2;
const pesoAvaliacaoSemestral = 3;
const pesoExameFinal = 5;

const notaFinal = (notaLaboratorio * pesoLaboratorio + 
                   notaAvaliacaoSemestral * pesoAvaliacaoSemestral + 
                   notaExameFinal * pesoExameFinal) / 
                  (pesoLaboratorio + pesoAvaliacaoSemestral + pesoExameFinal);

// Exibindo o resultado
alert("A nota final do estudante é: " + notaFinal.toFixed(2));

#exprof

let Lab=prompt('Digite Lab: ')
Lab = parseFloat(Lab);
let AVAL = prompt("Digite Avaliação Semestral: ")
AVAL = parseFloat(AVAL);
let Exame=prompt("Digite Exame Final: ")
Exame = parseFloat(Exame);

let media = (Lab*2+AVAL*3+Exame*5)/10
console.log("Sua média é: "+media);

let conceito;
if(media >=8.0 && media<=10.0){
    conceito ="A";
}else if(media>=7.0 && media<8.0){
    conceito ="B";
}else if(media>= 6.0 && media<7.0){
    conceito = "C";
}else if(media>= 5.0 && media < 6.0){
    conceito = "D";
}else{
    conceito = "E"
}
console.log("Seu conceito é: "+conceito);
```
